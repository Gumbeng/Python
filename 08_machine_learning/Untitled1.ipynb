```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = 0.2,
    stratify=y,
    random_state=0
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
unique_values_test, unique_values_cnt_test = np.unique(y_test, return_counts=True)
print(unique_values_test, unique_values_cnt_test)
print(unique_values_cnt_test/ y_test.size)
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train,
    test_size=0.2,
    stratify=y_train,
    random_state=0
)
X_train.shape, X_test.shape, X_val.shape,
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 모델 생성
max_depth = 4
tree = DecisionTreeClassifier(max_depth=max_depth, random_state=0)

# 학습
tree.fit(X_train, y_train)

# 검증
# 추론
pred_train = tree.predict(X_train)
pred_val = tree.predict(X_val)
# 검증
train_acc = accuracy_score(y_train, pred_train)
val_acc = accuracy_score(y_val, pred_val)
```

```python
print(f"max_depth: {max_depth}")
print("Train accuracy:", train_acc)
print("Validation accuracy:", val_acc)
```

```python
def sosu(n):
    if n == 1 or n == 0:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
```

```python
import time
primes = [True for _ in range(1000001)]
for i in range(2, 1000001):
    if primes[i] == True:
        for j in range(i*2, 1000001, i):
            primes[j] = False

T = int(input())

for i in range(T):
    n = int(input())
    s = time.time()
    partition = 0
    for num in range(2, n // 2 + 1):
        if primes[num] and primes[n - num]:
            partition += 1
    e = time.time()
    print(e-s)
    print(partition)

```
