```python
import requests
import json
custom_header = {
    'referer' : "https://www.kleague.com/record/team.do",
    'user-agent' : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36"
}
```

```python
def getResult():
    result = []
    url = "https://www.kleague.com/record/teamRank.do?leagueId=1&year=2023&stadium=all&recordType=rank"
    req = requests.get(url, headers = custom_header)

    if req.status_code == requests.codes.ok:
        print("연결 성공")
        stock_data = json.loads(req.text)
        for i in range(len(stock_data['data']['matchResultList'])):
            # print(stock_data['data']['teamRank'][i])
            result.append(stock_data['data']['matchResultList'][i])
        #matchResultList , teamRank
    else:
        print('접속 실패')
    return result
    
```

```python
a = getResult()
team_name = ['울산','포항','인천','제주','강원','수원FC','대구','서울','수원','김천','성남']

result = []
# 홈팀, 원정팀, 홈득, 원득, 승리, 비긴경기
for i in a:
    score = map(int, i["winTieLoss"].split(":"))
    win,loser,compare = score
    result.append([i["teamName"],i["otherTeam"],win, loser, compare])

for i in result:
    print(i)
    print("="*30)
```
